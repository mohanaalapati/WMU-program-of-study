```python
import pandas as pd

# Output from your code snippet
output = [
    "STAT 6620 - Applied Linear Models Credits: 3",
    "STAT 5860 - Computer Based Data Analysis Credits: 3",
    "STAT 5870 - Big Data Analysis Using Python Credits: 3",
    "STAT 6800 - SAS Programming Credits: 3",
    "CS 5430 - Database Systems Credits: 3",
    "CS 5610 - Advanced R Programming for Data Science Credits: 4",
    "CS 5821 - Machine Learning Credits: 3",
    "CS 6100 - Advanced Storage, Retrieval and Processing of Big Data Credits: 3",
    "STAT 6040 - Fundamentals of Epidemiology and Clinical Trials Credits: 3",
    "STAT 6500 - Statistical Theory I Credits: 3",
    "STAT 6600 - Statistical Theory II Credits: 3",
    "STAT 6640 - Applied Mixed Models Credits: 3",
    "CS 6030 - Studies in Computer Science Credits: 3",
    "CS 6260 - Advanced Parallel Computations Credits: 3",
    "CS 6310 - Advanced Design and Analysis of Algorithms Credits: 3",
    "CS 6430 - Database Management System Implementation Credits: 3",
    "CS 6530 - Data Mining Credits: 3",
    "CS 6820 - Advanced Artificial Intelligence Credits: 3",
    "CS 6821 - Information Retrieval Credits: 3",
    "STAT 6970 - Data Science Masters Project Credits: 4",
    "CS 6970 - Master’s Project Credits: 4"
]

# Initialize a list to store course information dictionaries
course_info_list = []

# Extract course information and append it to the list
for line in output:
    # Split the line by spaces
    parts = line.split()
    # Extract course code
    course_code = parts[0] + " " + parts[1]
    # Find the index where 'Credits:' appears
    credits_index = parts.index('Credits:')
    # Extract course name
    course_name_parts = parts[2:credits_index]
    # Remove leading hyphen if present
    if course_name_parts[0] == '-':
        course_name_parts = course_name_parts[1:]
    course_name = ' '.join(course_name_parts)
    # Extract credits
    credits = ' '.join(parts[credits_index + 1:])
    # Append the extracted information as a dictionary to the list
    course_info_list.append({"Course Code": course_code.strip(), "Course Name": course_name.strip(), "Credits": int(credits.strip())})

# Create a DataFrame from the list of dictionaries
course_df = pd.DataFrame(course_info_list)

# Print the DataFrame
print(course_df)
```
